```python
import pandas as pd
df=pd.DataFrame({"name":['A','B','C','D'],"Department":["CS","IT","CS","EC"]})
res=pd.get_dummies(df,columns=["Department"],)
print(res)
```

```python
import numpy as np
df=pd.DataFrame({"Score":[45,70,80,55]})
df["Result"] = np.where(df["Score"]>=60,"Pass","Fail")
df
```

```python
df=pd.DataFrame({"Score":[45,70,80,55]})
df["Grade"]=np.where(df["Score"]>=75,"A",np.where(df["Score"]>=60,"B","C"))
df
```

```python
import numpy as np
df=pd.DataFrame({"area":[1200,1500,1800,2200,2500],"price":[1500000,2200000,3100000,2800000,1800000]})
df["Grade"]=np.where(df["price"]<=2000000,"Low",np.where(df["price"]<=3000000,"Medium","High"))
df
```

```python
from sklearn.model_selection import train_test_split
df=pd.DataFrame({"Hours":[1,2,3,4,5,6],"Score":[50,55,65,70,80,90]})
x=df[["Hours"]]
y=df["Score"]
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=60)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```
