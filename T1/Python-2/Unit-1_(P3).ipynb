```python
df=pd.DataFrame({"maths":[80,85,90],
                "science":[78,82,88],
                "English":[70,75,80]})
df
```

```python
help(df.apply)
```

```python
print(df.apply(sum))
```

```python
result=df.apply(lambda x:x.mean())
print(result)
```

```python
result=df.apply(lambda x:x.mean(),axis=1,raw=True)
print(result)
```

```python
result=df.apply(lambda x:[x.sum(),x.mean()],axis=1,result_type='expand')
print(result)
```

```python
result=df.apply(lambda x:[x.sum(),x.mean()],axis=1)
print(result)
```

```python
result=df.apply(lambda x:x-x.mean(),axis=1,result_type="broadcast")
print(result)
```

```python
def add_bonus(x,bonus ):
    return x+bonus
result=df.apply(add_bonus,args=[5])
print(result)
```

```python
def add_bonus(x,bonus1,bonus2):
    return x+bonus1+bonus2
result=df.apply(add_bonus,args=[5,10])
print(result)
```
