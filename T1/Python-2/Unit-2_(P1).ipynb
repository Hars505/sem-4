```python
import networkx as nx
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4])
print(G)
G.add_edge(1,2)
G.add_edges_from([(1,1),(2,3),(3,4),(7,2)])
print(G)
```

```python
nx.draw_networkx(G,with_labels=False,node_color="green",edge_color="blue",node_size=100)
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3,4])
print(G)
G.add_edge(1,2)
G.add_edges_from([(1,1),(2,3),(3,4),(7,2)])
print(G)
```

```python
nx.draw_networkx(G,with_labels=True,node_color="brown",edge_color="blue",node_size=1000)
```
