```python
import pandas as pd
```

```python
df=pd.read_csv("placement.csv")
```

```python
df
```

```python
df.info()
```

```python
y=df["package"]
```

```python
x=df.drop("package",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import numpy as np
ypred = np.array(model.predict(x_test)) 
```

```python
ypred
```

```python
from sklearn import metrics
print(metrics.mean_squared_error(ypred,y_test))
print(metrics.mean_absolute_error(ypred,y_test))
import math
print(math.sqrt(metrics.mean_squared_error(ypred,y_test)))
print(metrics.r2_score(ypred,y_test))
```
