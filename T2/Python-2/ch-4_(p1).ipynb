```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
df = pd.read_csv("windpower.csv")
df
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
y=df["Wind speed (m/s)"]
x=df.drop("Wind speed (m/s)",axis=1)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
mse = []
mae = []
r2 = []
degrees=[3,4,5,6]
for i in range(3, 7):
    poly = PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(x)
    x_train, x_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.25, random_state=42)
    lr=LinearRegression()
    model = lr.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mse.append(metrics.mean_squared_error(y_test, y_pred))
    mae.append(metrics.mean_absolute_error(y_test, y_pred))
    r2.append(metrics.r2_score(y_test, y_pred))
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_xlabel('Polynomial Degree')
ax1.set_ylabel('Error Value (MSE/MAE)', color='tab:red')
ax1.plot(degrees, mse, label='MSE', marker='o', color='red', linestyle='--')
ax1.plot(degrees, mae, label='MAE', marker='s', color='orange', linestyle='-')
ax1.tick_params(axis='y', labelcolor='tab:red')
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.set_ylabel('R2 Score', color='tab:blue')
ax2.plot(degrees, r2, label='R2 Score', marker='^', color='blue', linestyle='-')
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.legend(loc='upper right')

plt.title('Model Performance vs Polynomial Degree')
plt.grid(True, alpha=0.3)
plt.show()
```
